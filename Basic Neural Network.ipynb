## Activation Function

### Sigmoid Activation function : It is a funtion that gives us a value between 0 and 1. This function is most suitable where we need probabilties as the output. Sigmoid function is used in classification problems where only 2 classes exist. Sigmoid funtion is widely used in output layer.

### tanh Activation function : tanh is a function that gives values between -1 and 1. It is expressed as the ratio of the difference between the exponential of the input value and the exponential of its negation to the sum of these exponentials. This function is applicable in hidden layer. 

### ReLU Activation function : ReLU function introduces the property of nonlinearity to a deep learning model and solves the vanishing gradient issue. It gives an output between 0 and infinity. This function is used in the hidden layer.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#y= 3x1 + 4x2
x1= 3
x2= 2
y_act=17
le= 0.01
error_= []
y_pred_= []

w1= np.random.randint(1,10)
w2= np.random.randint(1,10)

for epoch in range(1,21):
    y_pred= x1*w1 + x2*w2
    y_pred_.append(y_pred)
    
    error= (y_pred- y_act)**2
    error_.append(error)
    dedw1= 2*(y_pred- y_act)*x1
    dedw2= 2*(y_pred- y_act)*x2
    
    w1= w1- (le*dedw1)
    w2= w2- (le*dedw2)
    
    print(f'Iteration = {epoch}, y_actual =  {y_act}, y_pred = {y_pred}, error = {error}, w1 = {w1}, w2 = {w2}')
```

```python
plt.plot(error_)
plt.title('Error Graph')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.xticks(np.arange(0,21))
plt.show()
```

```python
plt.plot(y_pred_)
plt.title('y_pred Graph')
plt.xlabel('Epochs')
plt.ylabel('y_pred')
plt.xticks(np.arange(0,21))
plt.show()
```
